```python
import random
import numpy as np
from sklearn.datasets import load_iris


def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))


def group_distance(group, data):
    centroid = np.mean(data[group], axis=0)
    distances = [euclidean_distance(centroid, point) for point in data[group]]
    return np.mean(distances)


def evaluate_fitness(chromosome, data, k):
    groups = [[] for _ in range(k)]
    for i, label in enumerate(chromosome[:-1]):
        groups[label - 1].append(i)
    fitness = sum(group_distance(group, data) for group in groups)
    return fitness


def selection(population, data, k):
    fitness_values = [evaluate_fitness(chromosome, data, k) for chromosome in population]
    total_fitness = sum(fitness_values)
    probabilities = [fitness / total_fitness for fitness in fitness_values]
    selected_indices = random.choices(range(len(population)), probabilities, k=len(population))
    selected_population = [population[i] for i in selected_indices]
    return selected_population


def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 2)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2


def mutate(chromosome, k):
    mutation_point = random.randint(0, len(chromosome) - 2)
    new_label = random.randint(1, k)
    chromosome[mutation_point] = new_label
    return chromosome


def genetic_algorithm(data, k, population_size, num_generations):

    population = []
    for _ in range(population_size):
        chromosome = [random.randint(1, k) for _ in range(len(data))] + [k]
        population.append(chromosome)


    for _ in range(num_generations):
     
        selected_population = selection(population, data, k)

     
        new_population = []
        while len(new_population) < population_size:
           
            parent1, parent2 = random.choices(selected_population, k=2)

           
            child1, child2 = crossover(parent1, parent2)

           
            child1 = mutate(child1, k)
            child2 = mutate(child2, k)

           
            new_population.append(child1)
            new_population.append(child2)

        
        population = new_population


    best_chromosome = min(population, key=lambda chromosome: evaluate_fitness(chromosome, data, k))
    return best_chromosome[:-1]


iris = load_iris()
data = iris.data




best_chromosome = genetic_algorithm(data, k=3, population_size=150, num_generations=100)
print("Best chromosome:", best_chromosome)
```

```python
import numpy as np
import matplotlib.pyplot as plt



data = iris.data



labels = best_chromosome





plt.scatter(data[:, 0], data[:, 1], c=labels)

plt.xlabel('')
plt.ylabel('')
plt.title('Clustering Genetic')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans


iris = load_iris()
data = iris.data


kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)


labels = kmeans.labels_

# رسم نمودار خوشه‌بندی
colors = ['yellow', 'purple', 'blue']
markers = ['o', 's', '^']

for label in range(3):
    plt.scatter(data[labels == label, 0], data[labels == label, 1], c=colors[label], marker=markers[label])

plt.xlabel('')
plt.ylabel('Feature 2')
plt.title('Clustering Results')
plt.show()
```
